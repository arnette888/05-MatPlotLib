## Observations and Insights 

```python
# Dependencies and Setup
%matplotlib widget
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
# Merge two dataframes using an outer join
mouse_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
mouse_df.head()
```

```python
# Checking the number of mice.
# The unique method shows every element of the series that appears only once
unique = mouse_df["Mouse ID"].unique()
uniqueCount_1 = (len(unique));
uniqueCount_1
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_mouse_df = mouse_df.groupby(["Mouse ID", "Timepoint"]).size()
DuplicateMouse_df = duplicate_mouse_df.groupby(["Mouse ID", "Timepoint"]).size().reset_index(name="Time")
DuplicateMouse_df.head()
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate_mouse_df_2 = mouse_df.groupby(["Mouse ID","Timepoint","Drug Regimen","Sex","Age_months","Weight (g)","Tumor Volume (mm3)","Metastatic Sites" ]).size()
DuplicateMouse_df_2 = duplicate_mouse_df_2.groupby(["Mouse ID","Timepoint","Drug Regimen","Sex","Age_months","Weight (g)","Tumor Volume (mm3)","Metastatic Sites"]).size().reset_index(name="Time")
DuplicateMouse_df_2.head()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
# DuplicateMouse_df_3 = DuplicateMouse_df[DuplicateMouse_df.Timepoint != 0]
DuplicateMouse_df_3 = DuplicateMouse_df_2.query('Timepoint !=0')
DuplicateMouse_df_3.head()
```

```python
# Checking the number of mice in the clean DataFrame.
# The unique method shows every element of the series that appears only once
unique_DuplicateMouse_df_3 = DuplicateMouse_df_3["Mouse ID"].unique()
uniqueMouseCount_2 = (len(unique_DuplicateMouse_df_3));
uniqueMouseCount_2
```
